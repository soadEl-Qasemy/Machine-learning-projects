```python
def costfunction(x , y , w):
    J = 0
    m = y.size
    h = np.dot(x , w)
    J = (1 / (2 * m))* np.sum((h - y)**2)
    return J 
```

```python
data1 = np.loadtxt('data.txt', delimiter=',')
#print(data1)
y = data1[: , 4]
x = data1[: , :4]
m = len(y)
w =np.zeros(shape=(5 , 1))
new_x = np.ones(shape=(m , 5))
new_x[: ,  1:5] = x
#print(new_x)
J = costfunction(new_x , y , w)
print(f"Cost: {J}")
```

```python
def gradient(x , y , w , alpha , numb ):
    m = y.size
   
    J_prev = np.zeros(shape=(numb, 1))
    for i in range(numb):
        h = np.dot(x , w)
        error = h -y 
    
        #w -= (alpha / m) * np.dot(x.T, error)

        w[0] = w[0] - (alpha / m) * np.sum(error)
        w[1] = w[1] - (alpha / m) * np.sum(error * x[:, 1].reshape(m, 1))
        w[2] = w[2] - (alpha / m) * np.sum(error * x[:, 0].reshape((m, 1)))
        w[3] = w[3] - (alpha / m) * np.sum(error * x[:, 3].reshape(m, 1))
        w[4] = w[4] - (alpha / m) * np.sum(error * x[:, 4].reshape(m, 1))

        J_prev[i] = costfunction(x , y,w)
        #print(J_prev[i] )
    return w , J_prev 
    
```

```python
alpha = 0.01
iterations = 1000
```

```python
data1 = loadtxt('data.txt', delimiter=',')
#print(data1)
y = data1[: , 4].reshape(-1, 1)
x = data1[: , :4]
m = len(y)
w =np.zeros(shape=(5 , 1))
new_x = np.ones(shape=(m , 5))
new_x[: ,  1:5] = x
#print(new_x)
J = costfunction(new_x , y , w)
W, J_history = gradient(new_x, y, w, alpha, iterations)
print(J)
print("W : ", W)
```

```python
plot(arange(iterations), J_history)
xlabel('Iterations')
ylabel('Cost Function')
show()
```
